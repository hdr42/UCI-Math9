```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
N = 100
X = np.linspace(0,5,N)
Y = -2*X + 1
plt.plot(X,Y)
plt.show()
```

```python
# We manually create a data set with some errror.

N = 100
SamplingError = np.random.normal(0,1,X.shape[0]) # Mean: 0, Std: 1
Z = Y + SamplingError
plt.plot(X,Y)
plt.scatter(X,Z)
plt.show()
```

```python
# In reality all data we sample always contain error.
# Our goal is take the data and find a line that best represents the data.
plt.scatter(X,Z)
plt.show()
```

```python
# Using the closed form solution from lecture.
# N is total number of sample points we have. This was defined above.
S1 = np.sum(X)
S2 = np.sum(X * X)  # sum of the squares
SM = np.sum(X * Z)   # sum of x_i * y_i
SZ = np.sum(Z)
a = (1 / (N*S2 - S1*S1)) * (N * SM - S1 * SZ)
b = (1 / (N*S2 - S1*S1)) * (- S1 * SM + S2 * SZ)

plt.scatter(X, Z, alpha=0.3)
XX = np.linspace(0,5,200).reshape((-1,1))
ZZ = a * XX + b
plt.scatter(X, Z, alpha=0.2)
plt.plot(XX,ZZ,"r")
print("Approximation:\n","a = ",a,"\n b = ",b)
plt.show()
```

```python
# Using the sklearn package from lecture:
# N is total number of sample points we have. This was defined above.

from sklearn import linear_model
regr = linear_model.LinearRegression()
regr.fit(X.reshape((N,1)), Z.reshape((N,1)))
XXs = np.linspace(0,5,200).reshape((-1,1))
ZZs = regr.predict(XXs)
plt.scatter(X, Z, alpha=0.2)
plt.plot(XXs,ZZs,"r")
plt.show()
```
